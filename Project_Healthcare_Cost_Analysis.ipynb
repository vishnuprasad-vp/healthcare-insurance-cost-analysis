##  7. Data Preprocessing

We convert categorical values into numerical form
and create new features for better analysis.


# Convert Smoker Column

```python
df['is_smoker'] = df['smoker'].map({'yes': 1, 'no': 0})
```

Smoking status is converted into numeric format (1 = smoker, 0 = non-smoker).


# Create Age Groups

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[0, 18, 30, 45, 60, 100],
    labels=['0-18', '19-30', '31-45', '46-60', '60+']
)

df.head()
```

Age groups help analyze medical costs across different age ranges.


##  8. Exploratory Data Analysis (EDA)

We use visualizations to identify patterns and relationships.


# Distribution of Medical Charges

```python
plt.figure(figsize=(8,5))
sns.histplot(df['charges'], bins=30, kde=True)
plt.title("Distribution of Medical Charges")
plt.show()
```

Most people have lower medical charges,
while a few have very high expenses.


# Smoker vs Non-Smoker Charges

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='smoker', y='charges', data=df)
plt.title("Medical Charges: Smoker vs Non-Smoker")
plt.show()
```

Smokers have significantly higher medical charges than non-smokers.


# Charges by Age Group

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='age_group', y='charges', data=df)
plt.title("Medical Charges by Age Group")
plt.show()
```

Medical expenses increase as age increases.


# BMI vs Charges

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='bmi', y='charges', data=df, alpha=0.5)
plt.title("BMI vs Medical Charges")
plt.show()
```

Higher BMI often leads to higher medical costs.


# Correlation Heatmap

```python
plt.figure(figsize=(6,4))
sns.heatmap(
    df[['age','bmi','children','is_smoker','charges']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap")
plt.show()
```

Smoking has the strongest correlation with medical charges.
